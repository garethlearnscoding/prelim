```python
def partition(lis, L, R):
    i, j = L, L # 1 mark
    pivot = lis[R] # 1 mark
    while j < R:
        if lis[j] < pivot:
            lis[i], lis[j] = lis[j], lis[i] # 1 mark
            i += 1
        j += 1 # 1 mark
    lis[R], lis[i] = lis[i], lis[R] # 1 mark
    return i

def quicksort(lis):
    n = len(lis)
    def quicksort_helper(lis, L, R):
        if L < R:
            p = partition(lis, L, R) # 1 mark
            quicksort_helper(lis, L, p-1) # 1 mark
            quicksort_helper(lis, p+1, R)
    quicksort_helper(lis, 0, n-1) # 1 mark
    
lis = [9,2,3,5,1,4,6,8,7]
quicksort(lis)
print(lis)
```

```python
def insertionsort(lis):
    n = len(lis)
    for i in range(n): # 1 mark
        j = i
        while j > 0 and lis[j] < lis[j-1]: # 1 mark
            lis[j], lis[j-1] = lis[j-1], lis[j] # 1 mark
            j -= 1

lis = [9,2,3,5,1,4,6,8,7]
insertionsort(lis)
print(lis)
```
