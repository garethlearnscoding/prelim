```python
## Task 3.2
from datetime import datetime

def merge(A, B):
    ret = []
    while A and B:
        a_dob = datetime.strptime(A[0].birthdate,"%d%m%Y")
        b_dob = datetime.strptime(B[0].birthdate,"%d%m%Y")
        if  a_dob < b_dob :
            ret.append( A.pop(0))
        else:
            ret.append(B.pop(0))
    return ret + A + B

def sort_birthdays(L):
    if len(L) < 2:
        return L
    else:
        mid = len(L)//2
        return merge( sort_birthdays(L[:mid]), sort_birthdays(L[mid:]))
```

##### Marking scheme for non-optimised binary search
- standard bin search 
    - lb, ub , mid setup [1]
    - termination condition/base case [1]
    - case for 
        - < mid
        - \> mid 
        - update lb, ub [1]
    - == mid

    **Non-Optimised**
        -  sequential search left and right [2]

    **Optimised**
    - bin_search left[3]
    - bin_search right[3]

- return indexes [1]

```python
## O(N) Bin Search Max 6
def bin_search(L, target):
    lb = 0
    ub = len(L)-1 #[1]

    while lb <= ub: #[1]
        mid = (lb+ub)//2
        if L[mid].birthdate < target:
            lb = mid+1
        elif L[mid].birthdate > target:
            ub = mid -1 
        #[1] boundaries updated 
        else:
            # return mid 
            ## sequential search
            left_index = mid
            right_index = mid 
            ## search left [1]
            for i in range(mid-1, lb-1, -1):
                if L[i].birthdate == target:
                    left_index = i
                else:
                    break
            ## search right [1]
            for i in range(mid+1, ub+1):
                if L[i].birthdate == target:
                    right_index = i
                else:
                    break
            return (left_index, right_index) #[1]
            
    else:
        return (-1,-1)
```

```python
## O(log N) binary search with repeats
##[3] for standard binary search, 
# split into leftmost and rightmost search [6]
# continue search when index is found for leftmost and rightmost, boundaries updated [2]
# return indexes [1]
def search_birthday(L,target):
    def find_left(lb, ub, left_index=-1):
        if lb > ub:  
            return left_index #[1]
        mid = (lb + ub)//2 
        if target > L[mid].birthdate:
            return find_left(mid+1, ub, left_index) #[1] 
        elif target < L[mid].birthdate :
            return find_left(lb, mid-1, left_index) #[1] 
        else: ## continue search for leftmost
            left_index = mid
            return find_left(lb, mid-1, left_index) #[1] 

    def find_right(lb, ub, right_index=-1):
        if lb > ub: #
            return right_index #[1] 
        mid = (lb + ub)//2
        if target >  L[mid].birthdate:
            return find_right(mid+1, ub, right_index) #[1] 
        elif target < L[mid].birthdate :
            return find_right(lb, mid-1, right_index) #[1]
        else: ## continue search for rightmost 
            right_index = mid
            return find_right(mid+1, ub, right_index) #[1]
    
    lb = 0 
    ub = len(L)-1 #[1] 
 
    i = find_left(lb, ub) 
    j = find_right(lb, ub)

    return (i,j) #[1]
    
```
