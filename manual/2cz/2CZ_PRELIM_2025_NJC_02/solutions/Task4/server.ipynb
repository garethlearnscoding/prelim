```python
from flask import Flask, request, render_template, render_template_string
import sqlite3

app = Flask(__name__)

def get_all_competitor_names(): 
    conn = sqlite3.connect('task4.db')
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM competitor ORDER BY name") #[1] sql + cursor
    names = [row[0] for row in cursor.fetchall()] #[1] data retrieval
    conn.close()
    return names

@app.route('/query', methods=['POST']) #[1]
def query_competitor():
    name = request.form.get('competitor_name')

    conn = sqlite3.connect('task4.db')
    cursor = conn.cursor()

    # Find competitor ID [1]
    cursor.execute("SELECT id FROM competitor WHERE name = ?", (name,))
    row = cursor.fetchone()

    if row: #[2]
        competitor_id = row[0]
        # Get scores for that competitor
        cursor.execute("SELECT round, score FROM scores WHERE id = ?", (competitor_id,))
        scores = cursor.fetchall()
    else:
        scores = []

    conn.close()

    return render_template('results_completed.html', name=name, scores=scores) #[2]

''' 
[1] if result else no result
[1] for loop for round, score
[1] use of stylesheet elements
'''

# Route to display the form
@app.route('/') #[2] url mapping, render_template with data inputs
def form():
    return render_template('query_form_completed.html',  names = get_all_competitor_names())

'''
query form
[2] render competitors name using jinja
[1] query form 
[1] use of stylesheet elements
'''

if __name__ == '__main__':
    app.run()
```

```python
import sqlite3
conn = sqlite3.connect('task4.db')
cursor = conn.cursor()
cursor.execute("SELECT competitor.name, scores.score FROM scores \
INNER JOIN competitor ON scores.id = competitor.id \
WHERE scores.round = 3 ORDER BY scores.score DESC LIMIT 5") #[2]
results = [ {"name":name,"score":score} for name,score in cursor.fetchall() ] #[2]
conn.close()
with open("finalist.txt", "w") as final: #[1]
	final.write("[")
	di_str = ""
	for di in results:
		di_str += (f"{di},")
	final.write(di_str[:-1])
	final.write("]")
```
