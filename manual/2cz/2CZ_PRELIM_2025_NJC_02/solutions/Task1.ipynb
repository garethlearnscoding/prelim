```python
## Task 1.1
class Queue:
    def __init__(self, size):
        self.__array = [ None for _ in range(size)] #[1]
        self.__front = 0 # front of queue
        self.__back = 0 # insert at 
        self.__size = size
        self.__in_use = 0 # [1]  front, back, size, in_use

    def is_empty(self):
        return self.__in_use == 0 #[1]
    
    def is_full(self):
        return self.__in_use == self.__size #[1] edge case
    
    def enqueue(self, data):
        if self.is_full():
            print("Queue is Full")
            return #[1]
        self.__array[ self.__back]  = data
        self.__back = (self.__back + 1) % self.__size #[2] insert at correct position and update pointers
        self.__in_use +=1

    def dequeue(self):
        if self.is_empty(): #[1]
            print("Queue is Empty")
            return None
        else:
            ret = self.__array[self.__front] #[2] remove at correct position and update pointers
            self.__array[self.__front] = None
            self.__front = (self.__front + 1)%self.__size # wraparound
            self.__in_use -= 1
            return ret

    def __repr__(self):

        ret = [] # start -> .. -> back
        if self.is_empty(): #[1]
            return f"{ret}"
        
        cur = self.__front #[1]

        while len(ret) < self.__in_use: #[2] correct terminate condition for traversal
            ret.append(self.__array[cur])
            cur = (cur+1)%self.__size #[1]
       
        return f"{ret}"
```

```python
## Task 1.3

class Stack():
    def __init__(self, size): 
        self.__size = size
        self.__queue_1 = Queue(size) ## used for dequeing
        self.__queue_2 = Queue(size) ## used for enqueing # must be empty before enqueing
        self.__in_use = 0

    def __is_full(self): 
        return self.__in_use == self.__size
    
    def __is_empty(self):
        return self.__in_use == 0
    
    # -- [2]

    def __repr__(self): #[1]
        return repr(self.__queue_1)


    ''' Marking scheme for push and pop
    either implement addition operations in pop() or push 
    1 loop to empty + swap queue
    1 loop to empty + 1 loop to put back

    3 + 6 OR 6 + 3

    '''
    

    ''' push operation
    standard push using enqueue [3]
OR
    if additional code is used to maintain LIFO [6]
    '''
    
    def push(self, data):
        if self.__is_full():
            print("Stack is full") 
            return
        self.__queue_2.enqueue(data) #[1]
        self.__in_use += 1 #[1]
        while not self.__queue_1.is_empty(): #[2] #empty from other queue and enqueue 
            self.__queue_2.enqueue( self.__queue_1.dequeue())
        self.__queue_1 , self.__queue_2 = self.__queue_2, self.__queue_1 #[2] swap queue, enqueu must be done on empty queue

    ''' pop operation
    standard pop using dequeue [3]
OR
    if additional code is used to maintain LIFO [6]
    
    '''
    def pop(self):
        if self.__is_empty(): #[1]
            print("Stack is empty")
            return 
        ret = self.__queue_1.dequeue() #[1] order is LIFO
        self.__in_use -= 1 #[1]
        return ret
```
