```python
## Task 2.1 (3)
def read_data(filename):
    members = []
    with open(filename) as file:
        for line in file: #line seperator[1], field seperator[1]
            name,role,score = line.strip().split(',') #name,role,score
            members.append((name,role,(int(score)))) #[1]
    return members
```

```python
## Task 2.2a (3): Using sort
def get_roles(members):
    roles = {}
    for member in members: #[1]
        roles[member[1]] = roles.setdefault(member[1],[]) + [member] #[2]

## sorting mark (2) is given in 2.4
    # for role in roles:
    #     roles[role] = sorted(roles[role],key=lambda x: x[2], reverse=True) #key, descending
    return roles
```

```python
## Task 2.2b (3) using insertion sort
def get_roles(members): # name,role,score
    roles = {}
    for new_member in members: #[1]
        if new_member[1] in roles: #[1]

            ## insert in order ,(2) marks allocated in 2.4
            for i in range(len(roles[new_member[1]])): 
                if new_member[2] > roles[new_member[1]][i][2]:
                    roles[new_member[1]].insert(i, new_member) 
                    break
            else:
                roles[new_member[1]].append(new_member)
                
        else:
            roles[new_member[1]]=[new_member] #[1]
    return roles
```

```python
# Task 2.4a (6)
def form_squad(roles):
    # sorting done here or in 2.2
    for role in roles:
        roles[role] = sorted(roles[role],key=lambda x: x[2], reverse=True) #key, descending

    squads=[] 
    while all([roles[role] for role in roles]): 
        ## all the roles have at least one member
        #[1] iterate over all roles
        #[1] len(role) > 0
        #[1] exit loop if any role list is empty 
        squad = []
        for role in roles:
            # [2] remove role member with highest score
            
            
            c =  roles[role].pop(0) ## assume that list is sorted in descending order
            squad.append(c)
        squads.append(squad) #[1]
    return squads
            
```

```python
# Task 2.4b # list is not sorted
def form_squad(roles):
    squads=[]
    while all([roles[role] for role in roles]): ## all the roles have at least one member
    #[3]
        squad = []
        for role in roles:
            selected = max(roles[role], key=lambda x:x[2]) # find max item  in list
            roles[role].remove(selected) # remove max item from list
            #[2]
            squad.append(selected)
        squads.append(squad) #[1]
    return squads
            
```

```python
# Task 2.4c, list is sorted
def form_squad(roles):
    squads=[]

    num_roles = len(roles.keys())
    while True:
        #### all the roles have at least one member
        count=0
        for role in roles:
            if len(roles[role]) > 0:
                count+=1
        if count < num_roles: ## cannot form any more squads
            return squads
        ## [3]
        else:
            squad = []
            for role in roles: #[2] pick max score from role list
                c =  roles[role].pop(0)
                squad.append(c)
            squads.append(squad) #[1]

            
    return squads
```

```python
# Task 2.4c (list is not sorted)

def form_squad(roles):
    # sorting done here or in 2.2
    
    # determine the number of squads that can be formed
    # - find the role that has the least number of members
    number_of_squads = float("inf")
    for role in roles:
        roles[role] = sorted(roles[role],key=lambda x: x[2], reverse=True) #key, descending
        n = len(roles[role])
        if  n < number_of_squads:
            number_of_squads = n
    squads = [ [] for _ in range(number_of_squads)]
    for i in range(len(squads)):
        for role in roles:
            squads[i].append( roles[role].pop(0))
    return squads
```

```python
# Task 2.5 (3)
roles= get_roles(
    read_data("candidates.txt")
)
squads = form_squad(roles)
for i, squad in enumerate(squads,1):
    print( f"Squad {i}\t  Average Score:{round(sum(map(lambda x:x[2],squad))/len(squad),2)}")
    #[2] average score
    for member in squad: #[1]
        print(f"({member[0]},{member[1]})")
    print()
```
