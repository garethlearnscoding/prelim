# TASK 1 – Team Sorting and Matchup Generation

```python
def validate_student_code(code_list): #Max 4m, 5 possible marks
    """
    Validates student codes based on house code, numeric range, and uniqueness.
    Returns a list of valid codes and prints messages for invalid or duplicate entries.
    """
    valid_house_codes = ['BW', 'BB', 'HH', 'MR', 'MT']
    seen = []

    for code in code_list:
        housepart = code[:2] 
        if housepart not in valid_house_codes:    #1m House code validation
            print(f"Invalid house code: {code}")
            continue

        numberpart = code[2:]                     #1m Number validation
        try:
            if int(numberpart) >= 1 and int(numberpart) <= 99:
                if code not in seen:
                    seen.append(code)                    #1m - ensure no duplicates
        except:
            print(f"Invalid number part: {code}") #1m relevant rejection message
            continue
            
        
    return list(seen)
            
# Testing codes
lst_stud_codes = ["BW12", "MR5","HH88","MT3","BB8", "HH3","MR99","BW100","ZZ12","MR-5","MR", "MT3"] #1m - test with normal/abnormal/boundary data
print(validate_student_code(lst_stud_codes)) # Output: ['BW12', 'MR5', 'HH88', 'MT3', 'BB8', 'HH3', 'MR99']
```

```python
def extract_house_and_number(code):
    """
    Extracts the house code and numeric part from a student code.
    E.g., 'MR5' -> ('MR', 5)
    """
    house = code[:2].upper()
    number = int(''.join(filter(str.isdigit, code[2:])))
    return house, number

def insertion_sort_by_number(code_list):
    """
    Sorts a list of student codes based on:
    1. Student number (ascending)
    2. If tied, by house code (lexicographically)
    Uses custom Insertion Sort.
    """
    for i in range(1, len(code_list)):
        current = code_list[i]
        current_house, current_num = extract_house_and_number(current)
        j = i - 1
        while j >= 0:
            prev_house, prev_num = extract_house_and_number(code_list[j])
            if (
                prev_num > current_num or
                (prev_num == current_num and prev_house > current_house)
            ):
                code_list[j + 1] = code_list[j]
                j -= 1
            else:
                break
        code_list[j + 1] = current

# Testing codes
codes = ['BW12', 'MT3', 'HH3', 'MR99', 'HH88', 'MR5', 'BB8'] 
insertion_sort_by_number(codes)
print(codes)          # Output : ['HH3', 'MT3', 'MR5', 'BB8', 'BW12', 'HH88', 'MR99']
```

```python
# Task 1.3 – Read from file, validate, sort, split into teams
def process_student_codes(file_path):
    with open(file_path) as f:
        raw_codes = [line.strip() for line in f if line.strip()]

    valid_codes = validate_student_code(raw_codes)
    insertion_sort_by_number(valid_codes)

    team_blue = []
    team_red = []

    for code in valid_codes:
        house, number = extract_house_and_number(code)
        if number % 2 == 0:
            team_blue.append(code)
        else:
            team_red.append(code)

    return team_blue, team_red

# Testing codes 
print(process_student_codes("student_codes.txt"))
```

```python
# Task 1.4 – Generate Reverse Matchups
def generate_reverse_matchups(team_blue, team_red):
    rounds = max(len(team_blue), len(team_red))
    matchups = []

    for i in range(rounds):
        blue = team_blue[i] if i < len(team_blue) else "BYE"
        red = team_red[-(i + 1)] if i < len(team_red) else "BYE"
        matchups.append(f"Round {i + 1}: {blue} vs {red}")

    return matchups

#Testing codes

teamblue = ['BB8', 'BW12', 'HH88']
teamred = ['BW1', 'MT01', 'HH3', 'MT3', 'MR5', 'MR99']
generate_reverse_matchups(teamblue, teamred)
```

```python
# Sample test with student_codes.txt
test_file_path = "student_codes.txt"

team_blue, team_red = process_student_codes(test_file_path)
matchups = generate_reverse_matchups(team_blue, team_red)

print("Team Blue:", team_blue)
print("Team Red:", team_red)
print("\nMatchups:")
for line in matchups:
    print(line)
```
