```python
# Task 1.1
class Fish:
    #[2]
    #1-correct implementation of consume()
    #1-correct implementation of __str__()
    def __init__(self, name, size):
        self.name = name
        self.size = size
        
    def get_name(self):
        return self.name
    
    def get_size(self):
        return self.size
        
    def consume(self, fish):
        if self.size >= fish.size:
            return Fish(self.name, self.size + fish.size)
        else:
            return Fish(fish.name, self.size + fish.size)
            
    def __str__(self):
        return "Name:{:<20}Size:{:>4}".format(str(self.name), self.size)

f1 = Fish("Giant Trevally", 33)
f2 = Fish("Pollock", 59)
f3 = f1.consume(f2)
print(str(f3))
```

```python
import random
# Task 1.2
class Node:
    def __init__(self, data):
        self.data = data
        self.next_ptr = -1

# Task 1.2
class SortedLinkedListForFish:
    # Task 1.2a
    def __init__(self):
        # [3]
        # 1-correct initialisation of variables in __init__()
        # 2-correct implementation of next_free linked list.
        self.start = -1
        self.next_free = 0
        self.count = 0
        
        self.nodes = [] # Can always be replaced by two 1-D arrays
        for i in range(50):
            temp = Node(None)
            temp.next_ptr = i + 1
            self.nodes.append(temp)
        self.nodes[-1].next_ptr = -1
    
    # Task 1.2b
    def insert(self, fish):
        # [6]
        # 1-next_free slot successfully removed from next_free linked list and next_free variable is updated
        # 1-fish instance is successfully stored in the new slot
        # 1-position for fish node is searched and found based on size
        # 1-code to insert fish node in front of the linked list
        # 1-code to insert fish node in the middle and end of the linked list
        # 1-code increase count by 1
        if self.next_free != -1:
            # If link list is not empty and there is empty node available.
            new_slot = self.next_free
            self.next_free = self.nodes[self.next_free].next_ptr

            # To find the position in the linked list to insert the new node.
            # Remember the parent node of the new node as prev_index
            cur_index = self.start
            prev_index = -1
            while cur_index !=-1 and self.nodes[cur_index].data.size < fish.size:
                prev_index = cur_index
                cur_index = self.nodes[cur_index].next_ptr

            if prev_index == -1:
                # to insert new node at the head.
                self.start = new_slot
                self.nodes[new_slot].data = fish
                self.nodes[new_slot].next_ptr = cur_index                      
            else:
                # to insert new node at the middle or at the tail
                self.nodes[prev_index].next_ptr = new_slot
                self.nodes[new_slot].data = fish
                self.nodes[new_slot].next_ptr = cur_index #next item
            self.count += 1
        
    # Task 1.2c 
    def delete(self, fish_name):
        # [6]
        # 2-position of fish node based on fish_id_num is searched
        # 2-code to remove fish node from linked list when found
        # 1-code to return fish node to next free linked list
        # 1-code decrease count by 1
        if self.start != -1:
            # To find the node to be deleted
            # Remember the parent node of the node to be deleted as prev_index
            cur_index = self.start
            prev_index = -1
            while cur_index != -1 and self.nodes[cur_index].data.name != fish_name:
                prev_index = cur_index
                cur_index = self.nodes[cur_index].next_ptr
            
            if cur_index != -1:
                if prev_index == -1:
                    # delete the head node
                    self.start = self.nodes[cur_index].next_ptr
                    self.nodes[cur_index].data = None #reduncdant
                    
                    # return the deleted node (cur_index) back to 
                    # the next free linked list 
                    self.nodes[cur_index].next_ptr = self.next_free
                    self.next_free = cur_index
                else:
                    self.nodes[prev_index].next_ptr = self.nodes[cur_index].next_ptr
                    self.nodes[cur_index].data = None #reduncdant
                    
                    # return the deleted node (cur_index) back to 
                    # the next free linked list 
                    self.nodes[cur_index].next_ptr = self.next_free
                    self.next_free = cur_index
                self.count -= 1

    # Task 1.2d
    def display(self):
        # [1]
        # 1-code to output all the fishes information in the linked list
        cur_index = self.start
        while cur_index !=-1:
            print(str(self.nodes[cur_index].data))
            cur_index = self.nodes[cur_index].next_ptr

    # Task 1.3
    def pick2RandomFish(self):
        # [4]
        # 3-code to randomly identify two random fish nodes in linked list
        # 1-code to return two random fish instances in linked list
        
        if self.count > 1:
            # randomly picked 2 fishes from the LinkedList
            i = -1
            j = -1
            while i == j:
                i = random.randint(0, 5*self.count) % self.count
                j = random.randint(0, 5*self.count) % self.count
            if j < i:
                i, j = j, i
            lst = []
            cur_index = self.start
            index = 0
            while index <= j:
                if index == i:
                    lst.append(self.nodes[cur_index].data)
                elif index == j:
                    lst.append(self.nodes[cur_index].data)
                cur_index = self.nodes[cur_index].next_ptr
                index += 1
            return lst[0], lst[1]
        else:
            print("Not enough fish to pick.")
            return (None, None)
    
    # Task 1.3
    def preyTime(self):
        # [3]
        # 1-code to call pick2RandomFish to identify two fish and call consume with them
        # 1-code to delete both fish nodes
        # 1-code to insert back the winner
        fish1, fish2 = self.pick2RandomFish()
        winner = fish2.consume(fish1)
        self.delete(fish1.name)
        self.delete(fish2.name)
        self.insert(winner)
```

```python
# To test Task 1.3
def task1_3():
    # [4]
    # 1-open fishes.txt file and read each id_num and size in each line
    # 1-code to create a fish instance for each id_num and size and insert the fish instances into linked list
    # 1-code to call preyTime() till 1 fish is left
    # 1-display the fish and its size
    fish_lst = []
    f = open("fishes.txt", "r")
    line=f.readline()
    while line:
        id_num, size = line[:-1].split(',')
        fish_lst.append(Fish(id_num, int(size)))
        line=f.readline()
    f.close()
    
    ll = SortedLinkedListForFish()
    for fish in fish_lst:
        ll.insert(fish)
    while ll.count >1:
        ll.preyTime()
    ll.display()

for i in range(10):
    task1_3()
```
