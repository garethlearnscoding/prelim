```python
#Task 1.1 (no dictionary)
results = []

# file closes after with block
with open("votes.txt", "r") as file:
    lines = file.readlines()

for line in lines:
    moniker, artist = line.rstrip().split()
    # check if artist already in list
    found = False
    for row in results:
        if row[0] == artist:
            row[1] += 1
            found = True
            break

    if not found:
        results.append([artist, 1])

print(results)
```

```python
#Task 1.1 (using dictionary)
votes = {}
# file closes after with block
with open("votes.txt", "r") as file:
    lines = file.readlines()

for line in lines:
    moniker, artist = line.rstrip().split()
    if artist in votes:
        votes[artist] += 1
    else:
        votes[artist] = 1

# store in 2D array
results = []
for artist, count in votes.items():
    results.append([artist, count])

print(results)
```

```python
#Task 1.2
# Insertion Sort (descending order)
for i in range(1, len(results)):
    key = results[i]
    j = i - 1
    while j >= 0 and key[1] > results[j][1]:  # switched to > for desc order
        results[j + 1] = results[j]  # shift right
        j -= 1
    results[j + 1] = key  # insert key

print(results)

# Output most popular artists
highest = results[0][1]  # target vote count

print("Artists with highest vote count:")
i = 0
while i < len(results) and results[i][1] == highest:
    print(f"{results[i][0]}")
    i += 1
```

```python
def task1_3(arr, t):  # arr is sorted in descending order
    left = 0
    right = len(arr) - 1
    found_artists = []

    while left <= right:
        mid = (left + right) // 2

        if arr[mid][1] == t:  # a match is found
            found_artists.append(arr[mid][0])
            # find artists with same votes on the left
            for i in range(mid - 1, -1, -1):
                if arr[i][1] == t:
                    found_artists.append(arr[i][0])
                else:
                    break
            # find artists with same votes on the right
            for i in range(mid + 1, len(arr)):
                if arr[i][1] == t:
                    found_artists.append(arr[i][0])
                else:
                    break
            break  # match found, exit while loop

        elif arr[mid][1] > t:  # search right  (for arr in desc order)
            left = mid + 1

        else:   # search left
            right = mid - 1

    return found_artists

print(task1_3(results, 16))
print(task1_3(results, 2))
```
