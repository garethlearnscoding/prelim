Knowing how important data structures are in real life tech interviews in assessing problem solving skills, a Computing student decides to implement a hybrid data structure integrating binary search tree, linked list, stack and queue to manage players' game scores.  

Implement a binary search tree (BST) of stack or queue. Each BST node comprises the following:
- ds: a variable having the value 'S' (stack) or 'Q' (queue)
- score: game score
- data: stores player ids with the same score
  - if ds is 'S', data is a stack implemented using linked list
  - if ds is 'Q', data is a wraparound queue implemented using array
- left: pointer to left subtree
- right: pointer to right subtree


You may assume that a player only has one game score.  

Each stack or queue can store a maximum of 10 player ids with the same game score.
- a stack has a top pointer where insertion (push) and deletion (pop) happen, as well as a size attribute representing the number of items in the stack
- a queue has a front pointer where deletion (dequeue) happens and a rear pointer where insertion (enqueue) happens, as well as a size attribute representing the number of items in the queue  

Each data structure class has the following methods:
- BSTNode: constructor/initialiser
- BST: constructor/initialiser, insert(), search(), inorder()
- LinkedListNode: constructor/initialiser
- LinkedList: constructor/initialiser, insert()
- Stack: constructor/initialiser, push(), pop(), display()
- Queue: constructor/initialiser, enqueue, dequeue, display()

**Task 2.1**  
Write program code to create the classes BSTNode, BST, LinkedListNode, LinkedList, Stack and Queue with the associated attributes and methods. (16)

- appropriate code for queue (5)

```python
# Task 2.1
class BSTNode:
    def __init__(self, ds, key):
        self.ds = ds
        self.key = key
        if ds == 'S': # stack
            self.data = Stack()
        else: # queue
            self.data = Queue()
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, ds, key, player_id, node=None):
        if node is None: # first call
            node = self.root # start from root
        if not self.root: # empty bst
            self.root = BSTNode(ds, key) # insert as root
            if ds == 'S': # stack
                self.root.data.push(player_id)
            else: # queue
                self.root.data.enqueue(player_id)
        else: # non-empty bst
            if key < node.key: # go left
                if node.left: # recursively insert left
                    self.insert(ds, key, player_id, node.left)
                else: # insert as left leaf
                    node.left = BSTNode(ds, key)
                    if ds == 'S': # stack
                        node.left.data.push(player_id)
                    else: # queue
                        node.left.data.enqueue(player_id)
            elif key > node.key: # go right
                if node.right: # recursively insert right
                    self.insert(ds, key, player_id, node.right)
                else: # insert as right leaf
                    node.right = BSTNode(ds, key)
                    if ds == 'S': # stack
                        node.right.data.push(player_id)
                    else: # queue
                        node.right.data.enqueue(player_id)
            else: # same score
                if ds == 'S': # stack
                    node.data.push(player_id)
                else: # queue
                    node.data.enqueue(player_id)

    def search(self, key, node=None):
        if node is None: # first call
            node = self.root  # start from root
        if self.root is None: # empty bst
            return "Not found"
        if key == node.key: # found
            return node.data.display()
        elif key < node.key and node.left:
            return self.search(key, node.left)
        elif key > node.key and node.right:
            return self.search(key, node.right)
        else: # not found
            return "Not found"

    def inorder(self, node=None):
        if node is None: # first call
            node = self.root # start from root
        if self.root is None:
            return
        if node:
            if node.left:
                self.inorder(node.left)
            print(f"Key: {node.key}", end=' ')
            node.data.display()
            if node.right:
                self.inorder(node.right)

    # Task 2.4
    def maximum(self):
        if self.root is None:
            return "Tree is empty"

        curr = self.root
        while curr.right is not None:
            curr = curr.right

        max_score = curr.key
        player_ids = []

        if curr.ds == 'S':  # stack
            while curr.data.size != 0:
                player_ids.append(curr.data.pop())
        else:  # queue
            while not curr.data.size != 0:
                player_ids.append(curr.data.dequeue())

        return f"Highest score: {max_score}, Player ID(s): {', '.join(player_ids)}"


class LinkedListNode:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = LinkedListNode(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node


class Stack:
    def __init__(self):
        self.top = None
        self.size = 0

    def push(self, player_id):
        if self.size < 10:
            new_node = LinkedListNode(player_id)
            new_node.next = self.top
            self.top = new_node
            self.size += 1
        else:
            print("Stack is full. Cannot push more elements.")

    def pop(self):
        if self.top:
            popped = self.top.data
            self.top = self.top.next
            self.size -= 1
            return popped
        return None

    def display(self):
        current = self.top
        while current:
            print(current.data, end=" ")
            current = current.next
        print()


class Queue:
    def __init__(self):
        self.front = 0
        self.rear = 0
        self.size = 0
        self.array = [None] * 10

    def enqueue(self, player_id):
        if self.size < 10:
            self.array[self.rear] = player_id
            self.rear = (self.rear + 1) % 10
            self.size += 1
        else:
            print("Queue is full. Cannot enqueue more elements.")

    def dequeue(self):
        if self.size > 0:
            dequeued = self.array[self.front]
            self.front = (self.front + 1) % 10
            self.size -= 1
            return dequeued
        return None

    def display(self):
        if self.size == 0:
            print("Queue is empty")
            return
        index = self.front
        for _ in range(self.size):
            print(self.array[index], end=" ")
            index = (index + 1) % 10
        print()
```
