```python
# Task 2.1 Program code
class Node:
    def __init__(self):
        self.student_id = -1  # Default value
        self.student_name = ""  # Default value
        self.next_pointer = -1  # Initialised to -1 (null pointer)
```

```python
# Task 2.2 Program code
class LinkedList:
    def __init__(self):
        self.nodes_array = [Node() for _ in range(10)]  # Create 10 nodes
        self.start_pointer = -1  # Initially empty
        self.next_free_pointer = 0  # Points to first free node
        
        # Link all nodes into free space list
        for i in range(9):
            self.nodes_array[i].next_pointer = i + 1
        self.nodes_array[9].next_pointer = -1  # Last node points to -1
    
    # Task 2.3 Program code
    def insert_record(self, student_id, student_name):
        if self.next_free_pointer == -1:
            print("No free space available!")
    
        # Get the next free node
        new_node_ptr = self.next_free_pointer
        new_node = self.nodes_array[new_node_ptr]

        # Update free space list
        self.next_free_pointer = new_node.next_pointer

        # Set data in the new node
        new_node.student_id = student_id
        new_node.student_name = student_name
        new_node.next_pointer = -1

        # Insert into ordered linked list
        if self.start_pointer == -1:  # Empty list
            self.start_pointer = new_node_ptr
        else:
            current_ptr = self.start_pointer
            previous_ptr = -1

            # Traverse to find insertion point
            while current_ptr != -1 and self.nodes_array[current_ptr].student_id < student_id:
                previous_ptr = current_ptr
                current_ptr = self.nodes_array[current_ptr].next_pointer

            if previous_ptr == -1:  # Insert at start
                new_node.next_pointer = self.start_pointer
                self.start_pointer = new_node_ptr
            else:  # Insert in middle or end
                new_node.next_pointer = self.nodes_array[previous_ptr].next_pointer
                self.nodes_array[previous_ptr].next_pointer = new_node_ptr
    
   # Task 2.4 Program code
    def delete_record(self, student_id):
        if self.start_pointer == -1:
            print("Linked list is empty!")
            return
    
        current_ptr = self.start_pointer
        previous_ptr = -1

        # Traverse to find the record
        while current_ptr != -1 and self.nodes_array[current_ptr].student_id != student_id:
            previous_ptr = current_ptr
            current_ptr = self.nodes_array[current_ptr].next_pointer

        if current_ptr == -1:
            print("Record not found!")
            return
    
        # Delete the node and return to free space
        if previous_ptr == -1:  # Deleting first node
            self.start_pointer = self.nodes_array[current_ptr].next_pointer
        else:
            self.nodes_array[previous_ptr].next_pointer = self.nodes_array[current_ptr].next_pointer

        # Return node to free space list
        self.nodes_array[current_ptr].next_pointer = self.next_free_pointer
        self.next_free_pointer = current_ptr
        print("Record deleted successfully!")

    # Task 2.5 Program code
    def display_records(self):
        if self.start_pointer == -1:
            print("Linked list is empty!")
            return
    
        current_ptr = self.start_pointer
        while current_ptr != -1:
            node = self.nodes_array[current_ptr]
            print(f"Student ID: {node.student_id}, Name: {node.student_name}")
            current_ptr = node.next_pointer
```

```python
# Test for Task 2.2
linked_list = LinkedList()
print(f"Start Pointer: {linked_list.start_pointer}")
print(f"Next Free Pointer: {linked_list.next_free_pointer}")
print("Free Space List Pointers:")
for i, node in enumerate(linked_list.nodes_array):
    print(f"Node {i}: Next Pointer = {node.next_pointer}")
```
