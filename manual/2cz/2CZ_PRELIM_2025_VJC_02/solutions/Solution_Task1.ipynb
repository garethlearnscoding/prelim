```python
# Task 1.1

def task1_1(artefacts_list):
    
    valid_artefacts = []

    for i in range(len(artefacts_list)):
        artefact_id = artefacts_list[i][0]
        rating = artefacts_list[i][1]

        try:
            if not(artefact_id[0].isupper() and 1000 <= int(artefact_id[1:]) <= 9999):
                raise ValueError("ID does not start with uppercase letter and in the range between 1000 and 9999.")

            if not 1 <= int(rating) <= 100:
                raise ValueError("Rating must be an integer between 1 and 100.")

            valid_artefacts.append((artefact_id, int(rating)))
                
        except ValueError as error:
            print(f"Invalid artefact {i}: {error}")

        

    return valid_artefacts
```

```python
artefacts_list = [("A1234", 85),("B1234", 85),("C1234", 85),("B2567", 92),("A3891", 45),("C4012", 78),("B5678", 95),("D6543", 88),("E7890", 65),("A8123", 73),("C9012", 98),("B1567", 52)]
result = task1_1(artefacts_list)

print(f"Total number of valid artefacts: {len(result)}")

print("\nFirst 5 artefacts:")
for i in range(5):
    print(result[i])
```

```python
# Task 1.2

def quicksort(array):
    
    if len(array) <= 1:
        return array

    less_than = []
    more_than = []
    equal = []  

    pivot = array[0][1]

    for item in array:
        if item[1] < pivot:
            less_than.append(item)
        elif item[1] > pivot:
            more_than.append(item)
        else:
            equal.append(item)  

    return quicksort(more_than) + equal + quicksort(less_than)

def task1_2(artefacts_list):

    if not artefacts_list:
        return []
    
    return quicksort(artefacts_list)
```

```python
sorted_result = task1_2(result)

print("Sorted artefacts (by rating in descending order):")
for i in sorted_result:
    print(i)
```

```python
# Task 1.3

def binary_search_desc(array, target, low, high, results):

    if low > high:
        return

    mid = (low + high) // 2
    current_rating = array[mid][1]

    if current_rating == target:
        results.append(array[mid])

        left = mid - 1
        while left >= low and array[left][1] == target:
            results.append(array[left])
            left -= 1

        right = mid + 1
        while right <= high and array[right][1] == target:
            results.append(array[right])
            right += 1
        return

    elif current_rating < target:
        return binary_search_desc(array, target, low, mid - 1, results)
    else:
        return binary_search_desc(array, target, mid + 1, high, results)


def task1_3(artefacts_list, target_rating):

    sorted_artefacts = task1_2(artefacts_list)  
    results = []

    binary_search_desc(sorted_artefacts, target_rating, 0, len(sorted_artefacts) - 1, results)

    if results:
        return results
    else:
        return None
```

```python
rating = 85
print(f"Searching for artefact rating {rating}:")
if task1_3(result, rating):
    for artefact in task1_3(result, rating):
        print(f"ID: {artefact[0]}, Rating: {artefact[1]}")
else:
    print(f"No artefacts found with rating {rating}.")   
```
