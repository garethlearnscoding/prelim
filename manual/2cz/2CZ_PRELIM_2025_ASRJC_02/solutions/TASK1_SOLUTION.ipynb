```python
#Task 1.1
def encode_str(text, token):
    # validate input
    # check if token is detected
    if token in text:
        return f"Error: Text cannot contain the token {token}."
    
    # check for any existing digits
    for char in text:
        if char.isdigit():
            return "Error: Text cannot contain digits."
    
    # initialise variables
    i = 0
    result = []
    
    # loop through text
    while i < len(text):
        curr_char = text[i]
        count = 1
        
        # count consecutive characters
        while i + count < len(text) and text[i + count] == curr_char:
            count += 1
        
        # compress 3 or more consecutive characters
        if count >= 3:
            result.append(f"{token}{curr_char}{count}")
        else:
            result.append(curr_char * count)
        
        i += count
        
    return "".join(result)

# Test
print(encode_str("aaaaaaaaaaaaaabbcccd", "%"))  # output: %a14bb%c3d
print(encode_str("HHEEEEEEELLLPPPPPPPPPP", "$"))  # output: HH$E7$L3$P9
print(encode_str("!!!@@@@%%", "&"))  # output: &!3&@4%%
print(encode_str("abcdef%", "%"))  # contains token
print(encode_str("aaa123", "%"))  # contains digits
```

```python
#Task 1.2
def decode_str(encoded, token):
    i = 0
    result = []
    
    while i < len(encoded):
        if encoded[i] == token:
            # token found
            char = encoded[i + 1]
            count = ""  # to extract count
            j = i + 2
            
            # extract count
            while j < len(encoded) and encoded[j].isdigit():
                count += encoded[j]
                j += 1
            
            result.append(char * int(count))
            i = j
        else:
            # no token
            result.append(encoded[i])
            i += 1
    
    return "".join(result)

# Test
print(decode_str("%a14bb%c3d", "%"))
print(decode_str("HH$E7$L3$P10", "$"))
print(decode_str("&!3&@4%%", "&"))
```
