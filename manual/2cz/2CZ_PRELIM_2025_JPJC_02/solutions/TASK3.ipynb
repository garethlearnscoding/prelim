```python
def encrypt(code):
    encrypted = ""
    for ch in code:
        shifted = chr((ord(ch) - ord('a') + 3) % 26 + ord('a'))
        encrypted += shifted
    return encrypted
```

```python
def hashfunction(locker_id):
    return locker_id % 37
```

```python
def add(locker_id, hta):
    code = input(f"Enter codeword for locker ID {locker_id}: ")
    
    while True:    
        if code.isalpha() != True or code.islower() != True:
            print("Invalid codeword. Use only lowercase alphabetic characters.")
            code = input(f"Enter codeword for locker ID {locker_id}: ")
        else:
            break

    encrypted = encrypt(code)
    index = hashfunction(locker_id)

    # Linear probing
    original_index = index
    while hta[index] != None:
        index = (index + 1) % 37
        if index == original_index:
            print("Hash table is full.")
            return hta

    hta[index] = (locker_id, encrypted)
    return hta
```

```python
def search(hta):
    locker_id = int(input("Enter locker ID to access: "))
    code = input("Enter your codeword: ")
    
    encrypted = encrypt(code)
    index = hashfunction(locker_id)
    original_index = index

    while True:
        record = hta[index]
        if record is None:
            print("LOCKER NOT FOUND")
            break
            
        if record[0] == locker_id:
            if record[1] == encrypted:
                print("LOCKER IS OPEN")
                break
            else:
                print("INCORRECT CODE")
                break
            
        index = (index + 1) % 37
        
        if index == original_index:
            print("LOCKER NOT FOUND")
            break
```

```python
# Initialize hash table array with 37 slots
hta = [None] * 37

# Perform the test scenario
hta = add(36, hta)
hta = add(73, hta)
search(hta)
search(hta)
```
