```python
import csv
from typing import List

# -------------------------
# Task 1.1
# -------------------------
def readData(filename):
    """
    Reads treasures from a CSV file and returns a list of [name, value].
    CSV is expected to have 2 columns per row: name (str), value (int).
    """
    treasures: List[List] = []
        
    reader = open(filename, 'r')
    header = reader.readline()
    
    for line in reader:
        name, value = line.split(',')            
        treasures.append([name, int(value)])
            
    reader.close()
    
    return treasures

lst = readData("TASK1.csv")
print(lst)
```

```python
# -------------------------
# Task 1.2
# -------------------------
def quicksort(treasures):
    """
    Recursive quicksort of treasures by name (ascending, lexicographic).
    treasures: list of [name, value]
    """
    n = len(treasures)
    if n <= 1:
        return treasures[:]
    
    pivot_name = treasures[-1][0]
    less, equal, greater = [], [], []
    
    for item in treasures:
        name = item[0]
        if name < pivot_name:
            less.append(item)
        elif name > pivot_name:
            greater.append(item)
        else:
            equal.append(item)
    return quicksort(less) + equal + quicksort(greater)

lst = quicksort(lst)
print(lst)
```

```python
# -------------------------
# Task 1.3
# -------------------------
def binarySearch(treasures, name):
    """
    Iterative binary search by treasure name.
    treasures must be sorted by name ascending.
    Returns the value (int) if found; else -1.
    """
    lo, hi = 0, len(treasures) - 1
    
    while lo <= hi:
        mid = (lo + hi) // 2
        mid_name = treasures[mid][0]
        
        if mid_name == name:
            return treasures[mid][1]
        elif name < mid_name:
            hi = mid - 1
        else:
            lo = mid + 1
            
    return -1

print(binarySearch(lst, "Dragon Scale"))
```

```python
# -------------------------
# Helpers
# -------------------------
def display_inventory(treasures):
    if not treasures:
        print("(Inventory is empty)")
        return
    for name, value in treasures:
        print(f"{name} (Value: {value})")


# -------------------------
# Main Program (Task 1.4)
# -------------------------
treasures = readData("TASK1.csv")
current_list = treasures[:]

while True:
    print("\nWelcome to Your Treasure Inventory!")
    print("1. View Inventory")
    print("2. Sort Treasures by Name")
    print("3. Search Treasure by Name")
    print("4. Exit")

    choice = input("\nChoose option: ").strip()

    if choice == "1":
        display_inventory(current_list)

    elif choice == "2":
        current_list = quicksort(current_list)
        display_inventory(current_list)

    elif choice == "3":
        current_list = quicksort(current_list)  # ensure sorted
        target = input("Enter treasure name to search: ").strip()
        result = binarySearch(current_list, target)
        if result == -1:
            print("Treasure cannot be found")
        else:
            print(f"Found '{target}' worth {result} gold coins!")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid option. Please enter 1, 2, 3, or 4.")
```
