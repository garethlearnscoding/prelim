```python
#Task 3.1 - Create Node and Linked List class
class ContainerNode:

    def __init__(self, container_number, category_of_goods, cargo_weight): #1 mark - correct ContainerNode constructor & LinkedList constructor
        self.__container_number = container_number
        self.__category_of_goods = category_of_goods
        self.__cargo_weight = cargo_weight
        self.__next = None

    def get_container_number(self): #1 mark - correctly coded all getters and setters (including that of Linked List)
        return self.__container_number

    def get_category_of_goods(self):
        return self.__category_of_goods
    
    def get_cargo_weight(self):
        return self.__cargo_weight
    
    def get_next(self):
        return self.__next

    def set_container_number(self, container_number):
        self.__container_number = container_number

    def set_category_of_goods(self, category_of_goods):
        self.__category_of_goods = category_of_goods
        
    def set_cargo_weight(self, cargo_weight):
        self.__cargo_weight = cargo_weight
        
    def set_next(self, next):
        self.__next = next

class LinkedList:
    
    def __init__(self):
        self.__head = None
        
    def get_head(self):
        return self.__head
    
    def set_head(self, head):
        self.__head = head
    
    def insert_to_end(self, container_number, category_of_goods, cargo_weight):
        new_node = ContainerNode(container_number, category_of_goods, cargo_weight) #1 mark - Creating new ContainerNode for here and for insert_to_front method 
        if self.__head == None: #1 mark - Scenario 1 - empty Linked List
            self.__head = new_node
        else: #1 mark - Scenario 2 - not empty Linked List
            last_node = self.__head
            while last_node.get_next() != None: #iterating to the current last node
                last_node = last_node.get_next()
            last_node.set_next(new_node) #pointing the current last node to the new node (new node becomes the new last node)
    
    def insert_to_front(self, container_number, category_of_goods, cargo_weight):
        new_node = ContainerNode(container_number, category_of_goods, cargo_weight)
        if self.__head ==  None: #1 mark - Scenario 1 - empty Linked List
            self.__head = new_node
        else: #1 mark - Scenario 2 - not empty Linked List
            new_node.set_next(self.__head) #pointing the new node to the current head node
            self.__head = new_node #pointing the head pointer to the new node
    
    def delete_from_end(self):
        if self.__head == None: #1 mark - Scenario 1 - Linked List is empty
            return "Cannot delete - Empty"
        elif self.__head.get_next() == None: #1 mark - Scenario 2 - Linked List only has 1 node
            temp = self.__head
            self.__head = None #this will remove the node from the linked list as the head pointer is no longer pointing to it
            return temp #1 mark - returning the removed ContainerNode instance (together with Scenario 3 and delete_from_front Scenario 2) 
        else: #1 mark - Scenario 3 - Linked List has more than 1 nodes
            second_last_node = self.__head
            while second_last_node.get_next().get_next() != None: #iterating to search for the second last node
                second_last_node = second_last_node.get_next()
            temp = second_last_node.get_next() #second_last_node.get_next() will give the last node (the one to be deleted)
            second_last_node.set_next(None) #this will remove the last node from the linked list as the second_last_node is no longer pointing to it
            return temp 
        
    def delete_from_front(self):
        if self.__head == None: #1 mark - Scenario 1 - Linked List is empty
            return "Cannot delete - Empty"
        else: #1 mark - Scenario 2 - Linked List is not empty
            temp = self.__head
            self.__head = self.__head.get_next() #head pointer will point to the next node to be the new head node
            return temp #returning the removed ContainerNode instance 
```

```python
#Task 3.2 - Create Stack to inherit Linked List and add in 
class Stack(LinkedList): #1 mark - inherit from LinkedList
    
    def is_empty(self): #1 mark
        return self.get_head() == None #if self.head is None this will return True, otherwise it will return False
    
    def peek(self): #1 mark
        if self.get_head() == None:
            return "No containers"
        else:
            return self.get_head().get_container_number() #return the value of the top item in the stack without removing it from the stack
    
    def display(self): #1 mark
        if self.get_head() == None:
            print("No containers")
        else:
            curr_node = self.get_head()
            while curr_node != None:
                print(curr_node.get_container_number())
                curr_node = curr_node.get_next()
                
```
