A local media corporation is hosting a Top TV Artists Award. People can vote online for their favourite artist, out of 20 shortlisted ones. The voting results are stored in a file named `votes.txt`, where each line contains the voter's moniker followed by the artist they voted for, separated by a space. The following shows the first few records:
```
happypapaya Kieran_Tay
avgstudent Alarice_Soh
candycadet Amiyah_Kapoor
```

**Task 1.1**  
Write program code to read the data from `votes.txt` and store the number of votes each artist received in a 2D list `results`. Each row should contain the artist’s name and the corresponding vote count. (4)

- correct file input processing
- correct 2D array storage
- correct frequency update / appropriate data structure usage (2)

````python
# Task 1.1
infile = open('votes.txt', 'r')
lines = infile.readlines()
infile.close()
freqs = {}
for line in lines:
    line = line.strip().split()
    if line[1] in freqs: # existing
        freqs[line[1]] += 1
    else: # new
        freqs[line[1]] = 1

results = [] # initialise 2d array
for key, value in freqs.items():
    results.append([key, value])
print(results)
````

**Task 1.2**  
Write program code that applies the insertion sort to the 2D list `results`, sorting by the vote counts in descending order. Print `results` after sorting. (4)  
Write program code to identify and output the artist(s) with the highest number of votes. Note that there can be artists sharing the same number of votes. (2)

part(a) 
- correct insertion sort (2)
- descending 
- correct array index reference 
- correct sorted output  

part(b)
- correct logic for highest votes (2)
- correct output for highest voted

````python
# Task 1.2
def insertion_sort_descending(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j][1] < key[1]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key

# main
# sort results
insertion_sort_descending(results)
# output sorted results
print("Sorted Voting Results (Descending Order):")
for artist, count in results:
    print(f"{artist}: {count} votes")
    
# find highest voted artists
highest_votes = results[0][1] # first item is highest
highest_artists = [results[0][0]]
for i in range(1, len(results)):
    if results[i][1] == highest_votes: # same highest
        highest_artists.append(results[i][0])
    else: # early terminate
        break        
# output highest voted artists
print(f"Highest voted artists: {highest_artists}")
````

**Task 1.3**  
Write a function `task1_3` that uses binary search, modified to identify all artists that have received exactly t votes. Return a list containing these artists; otherwise return an empty list. Show the output for the cases where t is 16 and 2 respectively. (5)

- correct logic for array index
- correct logic for found list (2) 
- correct logic for not found empty list 
- correct output for found
- correct output for not found

````python
# Task 1.3 (iterative)
def binary_search_iterative(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid][1] == target: # found
            same = [arr[mid][0]] # first target
            i = mid - 1 # leftwards
            while arr[i][1] == target:
                same.append(arr[i][0])
                i -= 1
            j = mid + 1 # rightwards
            while arr[j][1] == target:
                same.append(arr[j][0])
                j += 1
            return same
        elif arr[mid][1] > target: # go left half
            left = mid + 1
        else: # go right half
            right = mid - 1
    return []  # not found

# main
print(binary_search_iterative(results, 16)) # found
print(binary_search_iterative(results, 2))  # not found
````
