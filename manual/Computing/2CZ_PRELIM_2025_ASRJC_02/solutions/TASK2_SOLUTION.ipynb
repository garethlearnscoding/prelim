```python
#Task 2.1
import random

def generate_runners():
    categories = ["Men's Marathon", "Women's Marathon", "Men's Half-Marathon", "Women's Half-Marathon"]
    runners = []  # populate runner data
    
    # store distinct phone numbers
    phone_nums = []
    while len(phone_nums) < 15_000:
        num = random.randint(8000_0000, 9899_9999)
        if num not in phone_nums:
            phone_nums.append(num)

    # create tuples
    for num in phone_nums:
        category = random.choice(categories)
        runners.append((num, category))

    # write to file
    # file closes after with block
    with open("runners.txt", "w") as file:
        for num, category in runners:
            file.write(f"{num},{category}\n")

    print("Generated", len(runners), "runners and saved file.")
    return runners

# Test
runners = generate_runners()
```

```python
#Task 2.2
import timeit

def merge_sort(arr):
    # base case
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])  # sorted left
    right = merge_sort(arr[mid:]) # sorted right

    return merge(left, right)

def merge(left, right):
    # helper function
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i][0] < right[j][0]:  # compare only phone numbers
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result = result + left[i:] + right[j:]  # store any remaining items at the rear
    return result

# Test
start_time = timeit.default_timer()
sorted_data = merge_sort(runners)
end_time = timeit.default_timer()

mergesort_time = end_time - start_time
print("Time taken:", mergesort_time, "seconds")

# Store data
with open("runners_sorted1.txt", "w") as file:
    for num, category in sorted_data:
        file.write(f"{num},{category}\n")
print("Data saved to file.")
```

```python
#Task 2.3
class Node:
    def __init__(self, phone, category):
        self.phone = phone
        self.category = category
        self.next = None

class OrderedLL:
    def __init__(self):
        self.head = None

    def insert(self, phone, category):
        new_node = Node(phone, category)

        # Insert at the front if LL is empty / data smaller than head value
        if self.head is None or phone < self.head.phone:
            new_node.next = self.head  # link new node to existing head
            self.head = new_node  # update head to this new node
        else:
            current = self.head
            # traverse to find the correct position to insert
            while current.next is not None and current.next.phone < phone:
                current = current.next  # move to next node

            # insert node
            new_node.next = current.next
            current.next = new_node

    def delete(self, phone):
        # check if empty
        if self.head is None:  # nothing to delete
            return False

        # check if first node is to be deleted
        if self.head.phone == phone:
            self.head = self.head.next  # skip first node
            return True

        # traverse linked list
        current = self.head
        while current.next is not None and current.next.phone != phone:
            current = current.next

        if current.next is not None:  # not at the end of LL, so found node to delete
            current.next = current.next.next  # skip the node
            return True

        return False  # did not find entry to delete

    def update(self, phone, new_category):
        current = self.head
        while current:
            if current.phone == phone:  # found match
                current.category = new_category  # update category
                return True  # exit while loop
            current = current.next  # move to next node
        return False

    def get_list(self):
        result = []
        current = self.head
        while current:
            result.append((current.phone, current.category))  # store tuple
            current = current.next
        return result

# Main
runners_LL = OrderedLL()  # instantiation

# Measure time for insertion
start_time = timeit.default_timer()  # start timer
for phone, category in runners:
    runners_LL.insert(phone, category)
end_time = timeit.default_timer()  # end timer

# Calculate time
OLL_time = end_time - start_time
print("Time taken:", OLL_time, "seconds")

# Extract list and store data
sorted_OLL = runners_LL.get_list()

with open("runners_sorted2.txt", "w") as file:
    for num, category in sorted_OLL:
        file.write(f"{num},{category}\n")
print("Data saved to file.")
```

```python
def bin_search(arr, target, left, right):
    if left > right:  # invalid search range
        return "Runner not found."
    
    mid = (left + right) // 2
    mid_val = arr[mid][0]  # extract phone number at middle index
    if target == mid_val:
        return arr[mid][1]  # found match, return category
    elif target < mid_val:
        return bin_search(arr, target, left, mid - 1)  # search left half
    else:
        return bin_search(arr, target, mid + 1, right)  # search right half

# Test
test1 = sorted_data[-1][0]  # last number
test2 = sorted_data[7495][0]  # near middle
test3 = 99999999  # not in list

for test in [test1, test2, test3]:
    result = bin_search(sorted_data, test, 0, len(sorted_data) - 1)
    print(f"Search for {test}: {result}")
```

```python
class CQueue:
    def __init__(self, capacity):
        self.data = [None] * capacity
        self.front = 0
        self.rear = 0  # points to the next available slot
        self.size = 0  # current number of items in queue
        self.capacity = capacity  # max capacity
    
    def is_empty(self):
        return self.size == 0
    
    def is_full(self):
        return self.size == self.capacity
    
    def enqueue(self, phone, category):
        if self.is_full():
            print("Queue is full. Cannot enqueue.")
            return
        
        self.data[self.rear] = (phone, category)  # store data
        self.rear = (self.rear + 1) % self.capacity  # update rear
        self.size += 1
        print("Enqueued:", phone, category)
    
    def dequeue(self):
        if self.is_empty():
            print("Queue is empty. Cannot dequeue.")
            return
        
        data = self.data[self.front]  # extra data from front
        self.data[self.front] = None  # optional: clear data in slot
        self.front = (self.front + 1) % self.capacity  # adjust front pointer
        self.size -= 1
        print("Dequeued:", data)
        return data
    
    def display(self):
        if self.is_empty():
            print("Queue is empty.")
            return
        
        print("Contents (front to rear):")
        curr = self.front
        for i in range(self.size):
            print(self.data[curr])
            curr = (curr + 1) % self.capacity

# Test
rq = CQueue(5)
print("Queue is empty:", rq.is_empty())
rq.enqueue(80000000, "Men's Marathon")
rq.enqueue(80000001, "Men's Marathon")
rq.enqueue(80000002, "Men's Marathon")
rq.enqueue(80000003, "Men's Marathon")
rq.enqueue(80000004, "Women's Marathon")
rq.enqueue(80000005, "Women's Marathon")
print("Queue is full:", rq.is_full())
rq.dequeue()
rq.dequeue()
rq.display()
```
