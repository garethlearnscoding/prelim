```python
# Task 4.1   
# Create the database and tables (6m)

import sqlite3

f1 = open('COURSES.db', 'w') #creates the db file
f1.close()

conn = sqlite3.connect('COURSES.db')

conn.execute('''CREATE TABLE Instructor (
    InstructorID TEXT PRIMARY KEY, 
    Name TEXT);''')

conn.execute('''CREATE TABLE Customer (
    CustomerID TEXT PRIMARY KEY, 
    Name TEXT);''')

conn.execute('''CREATE TABLE Course (
    CourseID TEXT PRIMARY KEY, 
    Name TEXT, 
    InstructorID TEXT, 
    StartDate TEXT, 
    Price INTEGER,
    FOREIGN KEY(InstructorID) REFERENCES Instructor(InstructorID));''')

conn.execute('''CREATE TABLE Registration (
    CustomerID TEXT, 
    CourseID TEXT,
    Paid INTEGER,
    PRIMARY KEY(CustomerID, CourseID)
    FOREIGN KEY(CustomerID) REFERENCES Customer(CustomerID),
    FOREIGN KEY(CourseID) REFERENCES Course(CourseID));''')

conn.commit()
conn.close()

# 1m: import sqlite3 and connection to database, incl. commit and close
# 1m: open and close file
# 1m: create Instructor and Customer tables
# 1m: create Course table with correct primary and foreign keys
# 1m: create Registration table with correct foreign keys
# 1m: ... and correct composite primary key
```

```python
#Task 4.2  
# Populate the 4 tables (5m)

import sqlite3

conn = sqlite3.connect('COURSES.db')

f1 = open("INSTRUCTOR.txt")

for line in f1:
    lst = line.strip().split(',')
    idno = lst[0]
    name = lst[1]

    conn.execute('''INSERT INTO Instructor (InstructorID, Name)
                    VALUES(?,?)''', (idno, name))

f1.close()

f2 = open("CUSTOMER.txt")

for line in f2:
    lst = line.strip().split(',')
    idno = lst[0]
    name = lst[1]

    conn.execute('''INSERT INTO Customer (CustomerID, Name)
                    VALUES(?,?)''', (idno, name))

f2.close()

f3 = open("COURSE.txt")

for line in f3:
    lst = line.strip().split(',')
    idno = lst[0]
    name = lst[1]
    instr = lst[2]
    date = lst[3]
    price = int(lst[4])

    conn.execute('''INSERT INTO Course (CourseID, Name, InstructorID, StartDate, Price)
                    VALUES(?,?,?,?,?)''', (idno, name, instr, date, price))

f3.close()

f4 = open("REGISTRATION.txt")

for line in f4:
    lst = line.strip().split(',')
    cust = lst[0]
    course = lst[1]
    paid = int(lst[2])
    
    conn.execute('''INSERT INTO Registration (CustomerID, CourseID, Paid)
                    VALUES(?,?,?)''', (cust, course, paid))
    
conn.commit()
conn.close()

# 1m: import sqlite3 and connection to database, incl. commit and close
# 1m: open and close at least one text file correctly
# 1m: extract data correctly and convert data types (at least one file)
# 1m: insert into at least one table correctly
# 1m: correctly done for all tables
```
