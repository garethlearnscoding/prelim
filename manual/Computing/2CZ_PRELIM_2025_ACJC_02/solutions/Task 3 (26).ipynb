```python
# task x.1

class Array:
    def __init__(self, n): # 1 mark for creating size and container
        self.container= []
        for i in range(n-1):
            self.container.append(["", i+1]) # 1 mark for populating container
        self.container.append(["", None])
        self.free = 0 # 1 mark
        
    def get_data(self, index): # 1 mark for get and set
        return self.container[index][0]
    
    def set_data(self, index, new_data):
        self.container[index][0] = new_data
    
    def get_pointer(self, index): # 1 mark for get and set
        return self.container[index][1]
    
    def set_pointer(self, index, new_pointer):
        self.container[index][1] = new_pointer

# Total: 4 marks        

# task x.2

class LinkedList:
    def __init__(self, array): # 1 mark
        self.array = array
        self.head = None
         
    def display(self):
        curr = self.head # 1 mark iteration
        while curr != None:
            print(self.array.get_data(curr)) # 1 mark printing data
            curr = self.array.get_pointer(curr)
    
    def add(self, new_data):
        if self.array.free == None: # 1 mark error message
            print("Error: No memory.")
        else:
            # 1 mark: create new node from the first node in free list
            new_node = self.array.free
            self.array.set_data(new_node, new_data)
            
            # 1 mark: move free list pointer to second node
            self.array.free = self.array.get_pointer(new_node)
            
            # 1 mark: iterate through linked list to find place to insert new node
            curr = self.head
            prev = None
            
            # 1 mark: which will be between prev and curr
            while curr != None and self.array.get_data(curr) < new_data:
                prev = curr
                curr = self.array.get_pointer(curr)
            
            # 1 mark: new_node's pointer is curr, prev's pointer is new_node
            self.array.set_pointer(new_node, curr)

            if prev != None:
                self.array.set_pointer(prev, new_node)
            else: # 1 mark: edge case: new node is at the head of linked list
                self.head = new_node
    
    def delete(self, to_delete):
        
        # 1 mark: iterate through linked list to find node to delete...
        curr = self.head
        prev = None
        
        # 1 mark: ... which will be curr, after prev
        while curr != None and self.array.get_data(curr) != to_delete:
            prev = curr
            curr = self.array.get_pointer(curr)
        
        if curr == None: #  1 mark: reached end of linked list without finding node
            print("Error: Node does not exist")
            
        else:
            #  1 mark: set prev's pointer to curr's pointer
            self.array.set_pointer(prev, self.array.get_pointer(curr))
            
            #  1 mark: set curr's data to ""
            self.array.set_data(curr, "")
            
            #  1 mark: set curr's pointer to head of freelist
            self.array.set_pointer(curr, self.array.free)
            
            #  1 mark: set freelist pointer to curr
            self.array.free = curr

# Initialising: 1 mark
# Display: 2 marks
# Add: 7 marks
# Delete: 7 marks
# Total: 18 marks

MyArray = Array(10) # 1 mark

AnimalList = LinkedList(MyArray) # 1 mark
PlantList = LinkedList(MyArray)

AnimalList.add("Cat") # 1 mark
PlantList.add("Rose")
AnimalList.add("Goat")
PlantList.add("Fern")
AnimalList.add("Horse")
PlantList.add("Vine")
AnimalList.delete("Cow") # 1 mark
PlantList.delete("Rose")
AnimalList.display() # 1 mark
PlantList.display()

# Testing: 5 marks

# total for this question: 27 marks
```
