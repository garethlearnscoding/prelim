```python
# Task 3.1

# Define node class

class Node:
    def __init__(self, team, left = None, right = None):
        self.team = team
        self.left_ptr = left
        self.right_ptr = right
        
    def __str__(self):
        return f"Node({self.team})"
        
```

```python
# Task 3.2

class TournamentTree:
    def __init__(self):
        self.root = None
        
    # Task 3.3
    def create_tree(self, num_of_teams):     
        leaf_nodes = []
        
        for i in range(num_of_teams):
            team = f"Team {chr(i+ord('A'))}"
            leaf_nodes.append(Node(team))
        
        count = 1
        nodes = leaf_nodes
        while len(nodes) > 1:
            new_nodes = []
            for i in range(0,len(nodes), 2):
                match = Node(f"Match {count}")
                match.left_ptr = nodes[i]
                match.right_ptr = nodes[i+1]
                new_nodes.append(match)
                count += 1
            
            nodes = new_nodes # move on to one level higher
            
        # len(nodes) == 1
        self.root = nodes[0]
        
        
    # Given code --------------------------------------------
    def level_order(self):
        '''
        print tree from root
        print out all the nodes of the same level in a row 
        '''
        self._level_order(self.root)
        
    def _level_order(self, node_list):
        '''
        Helper function for internal use
        print tree from node list
        '''
        # convert node_list to a list if it is not
        if not isinstance(node_list, list):
            node_list = [node_list]
        # Stop recursion if the list is empty
        if not node_list:
            return
        # define a list to collect node in next layer
        next_layer = []
        while node_list:
            node = node_list.pop(0) # Remove the first item
            print(node, end = ' ')
            if node.left_ptr:
                next_layer.append(node.left_ptr)
            if node.right_ptr:
                next_layer.append(node.right_ptr)
        print() # print new line after processing a layer
        self._level_order(next_layer)
        
    # -----------------------------------------------------
    
    # Task 3.4
    def pre_order(self):
        if self.root == None:
            return
        
        self._pre_order(self.root)
        print()
        
    def _pre_order(self, cur_node):
        print(cur_node.team, end = " ")
        if cur_node.left_ptr:
            self._pre_order(cur_node.left_ptr)
        if cur_node.right_ptr:
            self._pre_order(cur_node.right_ptr)
            
    def post_order(self):
        if self.root == None:
            return
        
        self._post_order(self.root)
        print()
        
    def _post_order(self, cur_node):
        if cur_node.left_ptr:
            self._post_order(cur_node.left_ptr)
        if cur_node.right_ptr:
            self._post_order(cur_node.right_ptr)
        print(cur_node.team, end = " ")
            
    # Task 3.5
    def search(self, team):
        if self.root == None:
            return False
        
        return self._search(self.root, team)
        
    def _search(self, cur_node, team):
        if cur_node == None:
            return False
        
        if cur_node.team == team:
            return True
        
        return self._search(cur_node.left_ptr, team) or self._search(cur_node.right_ptr, team)

    def calculate_matches(self, num_of_teams):
        if num_of_teams <= 1:
            return 0  # No matches needed for 0 or 1 team

        return num_of_teams - 1   
```

```python
# Task 3.3
t = TournamentTree()
t.create_tree(8)
t.level_order()
t.calculate_matches(8)
```

```python
# Task 3.4
print("Pre order")
t.pre_order()

print()
print("Post order")
t.post_order()

```
