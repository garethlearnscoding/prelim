```python
class PageNode:
    def __init__(self, url):
        self.url = url #string
        self.next = None #PageNode
        
class BrowserHistory:
    def __init__(self):
        self.head = None #PageNode

    def visit_page(self, url):
        new_node = PageNode(url)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next != None:
                current = current.next
            current.next = new_node

    def print_history(self):
        current = self.head
        print("Full Browsing History:")
        while current != None:
            print(current.url)
            current = current.next
```

```python
class BrowserNavigator:
    def __init__(self):
        self.current_page = "" 
        self.back_stack = [""] * 10
        self.forward_stack = [""] * 10
        self.back_top = -1
        self.forward_top = -1

    def go_back(self):
        if self.back_top != -1:
            self.forward_top += 1
            self.forward_stack[self.forward_top] = self.current_page

            self.current_page = self.back_stack[self.back_top]
            self.back_top -= 1
        else:
            print("No pages to go back to.")

    def go_forward(self):
        if self.forward_top != -1:
            self.back_top += 1
            self.back_stack[self.back_top] = self.current_page

            self.current_page = self.forward_stack[self.forward_top]
            self.forward_top -= 1
        else:
            print("No pages to go forward to.")

    def visit_new_page(self, url):
        if self.back_top < 9:
            self.back_top += 1
            self.back_stack[self.back_top] = self.current_page
        else:
            print("Back stack is full.")

        self.current_page = url
        self.forward_top = -1
```

```python
# --- Create instances ---
history = BrowserHistory()
navigator = BrowserNavigator()

# --- Visit 3 pages ---
urls = ["jpjc.coursemology.org", "jpjc.moe.edu.sg", "web.whatsapp.com"]
for url in urls:
    navigator.visit_new_page(url)
    history.visit_page(url)

# --- Go back twice ---
navigator.go_back()
navigator.go_back()

# --- Go forward once ---
navigator.go_forward()

# --- Output the current page ---
print("Current Page:", navigator.current_page)

# --- Output full browsing history ---
history.print_history()
```
