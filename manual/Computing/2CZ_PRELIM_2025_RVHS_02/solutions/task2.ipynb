```python
########
# AETS #
# TORY #
# ISHE #
# OMEA #
########

#######
# 123 #
# 8C4 #
# 765 #
#######


board = [["A","E","T","S"],["T","O","R","Y"],["I","S","H","E"],["O","M","E","A"]]
valid_moves = [[2,1,4,1,4,2], [2,1,7,4,4], [2,1,1,4,4,5],[0,3,8,7,4,4],[1,2,8,6,5], [2,0,4]]
invalid_moves = [[2,1,8,8], [2,1,7,6], [2,1,4,4,4],[0,3,2,2,2],[1,2,3,7,7,7,7]]

# Task 2.1
def valid_dir(board, curr_pos, direction): 
    # [4]
    # 1-correct change in coordinate based on direction given
    # 2-able to handle out of bound error
    # 1-correctly implemented
    move_by = {1:(-1,-1), 2:(-1,0), 3:(-1,1), 4:(0,1), 5:(1,1), 6:(1,0), 7:(1,-1), 8:(0,-1)}
    dr, dc = move_by[direction]
    fr, fc = dr+curr_pos[0], dc+curr_pos[1]
    if fr < 0 or fr > len(board)-1 or fc < 0 or fc > len(board[0])-1:
        return False
    else:
        return True

# Task 2.2
def result_word(board, move):
    # [6]
    # 1-code to identify original pos and the steps in move
    # 1-code to move the pos based on steps in move
    # 1-code to traverse all the characters in board based on move
    # 1-code to store all letters visited in a variable
    # 1-code to use valid_dir to determine an invalid move
    # 1-correct output
    move_by = {1:(-1,-1), 2:(-1,0), 3:(-1,1), 4:(0,1), 5:(1,1), 6:(1,0), 7:(1,-1), 8:(0,-1)}
    pos = move[:2]
    result=board[pos[0]][pos[1]]
    for direction in move[2:]:
        if valid_dir(board, pos, direction):
            dr, dc = move_by[direction]
            pos[0], pos[1] = dr+pos[0], dc+pos[1]
            result+=board[pos[0]][pos[1]]
        else:
            return None
    return result

for move in valid_moves + invalid_moves:
    print(result_word(board, move))
```

```python
score_board = [[1,2,3,1],[2,2,3,1],[1,2,2,1],[1,4,2,1]]
# Task 2.3
def result_score(score_board, move):
    # [4]
    # 1-code to move in the score_board based on steps in valid_move
    # 1-code to traverse all the number in board based on move
    # 1-code to sum up all the numbers visited in a variable
    # 1-correct output    
    move_by = {1:(-1,-1), 2:(-1,0), 3:(-1,1), 4:(0,1), 5:(1,1), 6:(1,0), 7:(1,-1), 8:(0,-1)}
    pos = move[:2]
    result=score_board[pos[0]][pos[1]]
    for direction in move[2:]:
        dr, dc = move_by[direction]
        pos[0], pos[1] = dr+pos[0], dc+pos[1]
        result+=score_board[pos[0]][pos[1]]
    return result

for move in valid_moves:
    print(result_score(score_board, move))
```

```python
# Task 2.4
def word_scores(board, score_board, valid_moves):
    # [3]
    # 1-code to use result_word() and result_score()
    # 1-code to find both the word and its score based in moves
    # 1-code to store all the word and its score in a 2D list
    result = []
    for move in valid_moves:
        word = result_word(board, move)
        if word:
            word_score = result_score(score_board, move)
            result.append([word, word_score])
    return result
print(word_scores(board, score_board, valid_moves))
```
