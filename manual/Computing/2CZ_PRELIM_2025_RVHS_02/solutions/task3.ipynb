```python
# Task3.1
def convert10(num7_str):
    if len(num7_str) < 2:
        return int(num7_str)
    else:
        return convert10(num7_str[:-1])*7 + int(num7_str[-1])

def decipher(alien_code):
    result = ""
    for i in range(0, len(alien_code), 3):
        b10 = convert10(alien_code[i:i+3])
        result += chr(b10)
    return result

def display_alien_words(alien_codes):
    print("{:<12}{:<}".format("WORD", "CODE"))
    for alien_code in alien_codes:
        print("{:<12}{:<}".format(decipher(alien_code), alien_code))
        
display_alien_words(alien_codes)
```

```python
# Task3.3
class AlienKnowledgeDataBase2():
    
    def hashfunction(self, key):
        hashvalue = 0
        for char in key:
            hashvalue += ord(char)
        return hashvalue%103
    
    def __init__(self, known_alien_word_list):  
        # assume that the hash table will never be full.
        self.knowledgebase = [None]*103
        for known_alien_word in known_alien_word_list:
            hashvalue = self.hashfunction(known_alien_word)
            while self.knowledgebase[hashvalue]:
                hashvalue = (hashvalue+1) % 103
            self.knowledgebase[hashvalue] = known_alien_word
            
            
    def search(self, alien_word): # O(1) time
        hashvalue = self.hashfunction(alien_word)
        while self.knowledgebase[hashvalue] and self.knowledgebase[hashvalue]!=alien_word:
            hashvalue = (hashvalue+1) % 103
        if self.knowledgebase[hashvalue]:
            return True
        else:
            return False
        

def task3_3():
    akd = AlienKnowledgeDataBase2(known_alien_word_list)
    result = []
    for alien_word in alien_words_100:
        if not akd.search(alien_word):
            result.append(alien_word)
    print("There are " + str(len(result)) + " unknown alien words found:")
    print(result)

task3_3()
```

```python
def has_duplicates(lst):
    seen = set()
    for item in lst:
        if item in seen:
            return True
        seen.add(item)
    return False

print(has_duplicates(english_words_100), len(english_words_100))
print(has_duplicates(alien_words_100), len(alien_words_100))
```
